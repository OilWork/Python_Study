## plot() 
- kind 매개변수에 지정한 값에 따라 다양한 그래프를 그릴 수 있다.
- kind : 그래프 종류 지정
    - 'line' : line plot (default)
    - 'bar' : vertical bar plot
    - 'barh' : horizontal bar plot
    - 'hist' : histogram
    - 'box' : boxplot
    - 'kde' : Kernel Density Estimation plot
    - 'pie' : pie plot
    - 'scatter' : scatter plot

- pd.date_range() 
    - 일정 간격으로 증/감하는 특정 범위의 날짜를 가지는 Index(DatetimeIndex)를 생성
        - 시계열 데이터의 행 index를 날짜시간으로 만들때 사용

```python
# (시작날짜, 끝날짜)
# pd.date_range('2022-06-01', '2022-06-10')
pd.date_range('2022-06-01', '2022-10-10', periods=5 )
```

```python
# (시작날짜, freq='증감간격기준'. period=개수)
# freq:Y-년도, M - 월별, D(기본) - 날짜 (해당 기준의 마지막일시)
#      뒤에 S (해당 기준의 시작일시) -YS, MS, DS
# pd.date_range('2022-01-01', freq='M', periods=5)
pd.date_range('2022-01-01', freq='MS', periods=5)
```

```python
# freq 문자열 앞에 숫자붙이면 그 숫자의 기간만큼 증감
# pd.date_range('2022-01-01', freq='5MS', periods=5)
pd.date_range('2022-01-01', freq='3YS', periods=5)
```

```python
# 평균이 5이고 표준편차가 3인 정규분포를 따르는 난수 생성
# size: (10,3) shape
value = np.random.normal(5,3,size=(10,3))
index = pd.date_range('2022-06-01', freq='D',periods=10)
# index
df = pd.DataFrame(value, index=index, columns=['no1','no2','no3'])
```

```python
# index명이 연속된 일시(DatetimeIndex)의 DateFrame => 시계열 데이터. (일시 변화에 따른 어떤 값의 변화를 가지는 데이터셋)
df
```

```python
# series/dataframe.plot() : 선그래프
# X: index명, Y:값
df['no1'].plot(figsize=(10,5), title='제목', ylabel='no1 value', xlabel='날짜')  # 
plt.show()
```

```python
df.plot(grid=True, figsize=(10,5))
# plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.title('제목')
plt.ylabel('no1 value')
plt.xlabel('날짜')
plt.plot(df['no1'])
plt.plot(df['no2'])
plt.plot(df['no3'])
plt.show()
```

```python
# 여러개 컬럼의 값들을 하나의 axes에 그리기
# dataframe.plot메소드()
df[['no1','no2']].plot(figsize=(15,5))
plt.show()
```

matplotlib를 이용해 여러 설정들 가능

```python
# series.plot() : plot 메소드가 아래작업을 대신 해준다
plt.figure(figsize=(10,5))
plt.title('제목')
plt.ylabel('no1 value')
plt.xlabel('날짜')

df['no1'].plot()
plt.show()
```

```python
df.reset_index(drop=True).plot()   # x축: index명
plt.legend(loc='upper left', bbox_to_anchor=(1,1))
plt.show()
```

## 막대 그래프
- index가 무슨 값인지를 가리키는 축으로 사용된다.
    - index: 수량을 세려는 대상
    - value: 수량

```python
tips = pd.read_csv('data/tips.csv')
tips.head()
```

```python
# 성별 수량
tips['sex'].value_counts().plot(kind='bar')
plt.show()
```

```python
tips['sex'].value_counts().plot.bar()
plt.show()
```

```python
# 요일의 개수를 수평막대그래프
# tips['day'].value_counts().plot(kind='barh')
tips['day'].value_counts().plot.barh()
plt.show()
```

```python
tips.head()
```

- 두개의 분류 별로 그리기
    - 여러개의 컬럼일 경우 수평 누적 막대그래프를 그린다.


```python
# 요일 성별 인원수 합계 막대그래프 -> pivot_table로 정렬 후에 막대그래프로 그리기
result_df = tips.pivot_table(index='day', columns='sex', values='size', aggfunc='sum')
result_df.plot(kind='bar',figsize=(7,7))
```

```python
tips.pivot_table(index='day', columns='smoker', values='tip',aggfunc='mean')
```

```python
# 요일별, 흡연여부의 팁의 평균
tips.pivot_table(index='day', columns='smoker', values='tip',aggfunc='mean').plot(kind='bar', figsize=(7,7), rot=45)
plt.show()
```

```python
# index: 수량을 셀 groupby 컬럼, columns: 막대그래프를 나눠서 볼 groupby 컬럼
tips.pivot_table(index='day', columns='smoker', values='tip',aggfunc='mean').plot(kind='barh', figsize=(7,7), rot=45)
plt.show()
```

```python
tips.pivot_table(index='smoker', columns='day', values='tip',aggfunc='mean').plot(kind='bar', figsize=(7,7), rot=45)
plt.show()
```

## Boxplot (상자그래프)

```python
tips['total_bill'].plot(kind='box', figsize=(7,7))
```

극단치 : 정상범위를 넘어선 값들. 이상치(Outlier-극단치, 오류값) 중에 하나
- boxplot에서 극단치를 계산하는 방법.
1. IQR(Inter Quantile Range) 계산
    - 3분위값-1분위값(중간 50%의- 25%~75%의 크기
2. 정상범위 이하의 극단치(극단적으로 작은값): 1분위값 - IQR * 1.5 
3. 정상범위 이상의 극단치(극단적으로 큰값):  3분위값 + IQR * 1.5

1.5: 조절할 수 있는 값 => whis 겂
정상범위를 늘리고 싶을 경우 1.5 보다 큰값을 곱하면 된다.
정삼범위를 줄이고 싶을 경우 1.5 보다 작은값을 곱하면 된다.

```python
quantile = tips['total_bill'].quantile(q=[0.25, 0.5, 0.75])
IQR = quantile[0.75] - quantile[0.25]
print("IQR", IQR)
whis = 1.5
o1 = quantile- IQR*
```
